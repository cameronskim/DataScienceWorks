```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
from tabulate import tabulate

df_reg=df[['Original Ask Amount', 'Total Deal Amount', 'Valuation Requested', 'Number of Sharks in Deal']].dropna()
```

```python
X_simple=df_reg[['Original Ask Amount']]
Y=df_reg['Total Deal Amount']

X_simple=sm.add_constant(X_simple)

model=sm.OLS(Y,X_simple).fit()

Y_pred_simple=model.predict(X_simple)

simple_results=[
    ["Intercept (ß₀)", model.params[0], model.pvalues[0]],
    ["Original Ask Amount (ß₁)", model.params[1], model.pvalues[1]]
]

print("\n### Simple Linear Regression Results ###")
print(tabulate(simple_results, headers=["Coeficient", "Value", "P-Value"], tablefmt="grid"))

mae_simple = mean_absolute_error(Y,Y_pred_simple)
mse_simple = mean_squared_error(Y,Y_pred_simple)
rmse_simple = np.sqrt(mse_simple)
r2_simple = r2_score(Y,Y_pred_simple)

simple_Metrics = [
    ["Mean Absolute Error (MAE)", mae_simple],
    ["Mean Squared Error (MSE)", mse_simple],
    ["Root Mean Squared Error (RMSE)", rmse_simple],
    ["R-Squared (R²)", r2_simple]
]

print("\n### Simple Linear Regression Metrics ###")
print(tabulate(simple_Metrics, headers=["Metric", "Value"], tablefmt="grid"))

plt.figure(figsize=(8,6))
sns.regplot(x=df_reg['Original Ask Amount'], y=df_reg['Total Deal Amount'], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.xlabel("Original Ask Amount ($)")
plt.ylabel("Total Deal Amount ($)")
plt.title("Simple Linear Regression: Original Ask Amount vs Total Deal Amount")
plt.show()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from tabulate import tabulate

selected_features=[
'Original Ask Amount','Industry','Pitchers Gender','Multiple Entrepreneurs'
]

df_model=df[selected_features+['Got Deal']].dropna().copy()

df_model['Got Deal'] = df_model['Got Deal'].astype(int)

numeric_cols=['Original Ask Amount', 'Multiple Entrepreneurs']
for col in numeric_cols:
  df_model[col]=pd.to_numeric(df_model[col], errors='coerce')

df_model=pd.get_dummies(df_model, columns=['Industry', 'Pitchers Gender'], drop_first=True)

Y = df_model['Got Deal']
X = df_model.drop('Got Deal', axis=1)
X = sm.add_constant(X)
```

```python
X=df_model.drop(columns=['Got Deal'])
y=df_model['Got Deal']

X=X.apply(pd.to_numeric, errors='coerce')

X=sm.add_constant(X)

X=X.astype(float)

logit_model=sm.Logit(y,X)
result=logit_model.fit()

ll_null=result.llnull
ll_model=result.llf
pseudo_r2=1-(ll_model/ll_null)

coef_df=pd.DataFrame({
    "Features":X.columns,
    "Coefficients":result.params,
    "Odds Ratio":np.exp(result.params),
    "P-Value":result.pvalues,
})

coef_table=tabulate(coef_df, headers="keys", tablefmt="grid", floatfmt=".4f")

y_pred=result.predict(X)
y_pred=(y_pred>0.5).astype(int)

conf_matrix=confusion_matrix(y,y_pred)
class_report=classification_report(y,y_pred,output_dict=True)

#Convert classification report to tabulated format
class_report_table=tabulate(
    [[metric, f"{values['precision']:.2f}", f"{values['recall']:.2f}", f"{values['f1-score']:.2f}"]
     for metric, values in class_report.items() if metric in ['0', '1']],
    headers=["Class", "Precision", "Recall", "F1-Score"],
    tablefmt="grid"
)

print("\n=== Logistic Regression Coeeficients & Odds Ratios ===\n")
print(coef_table)

print("\n=== Pseudo R² ===\n")
print(pseudo_r2)

print("\n=== Classification Report ===\n")
print(class_report_table)

plt.figure(figsize=(6,5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", cbar=False, linewidths=1, linecolor="black",
            xticklabels=["Predicted No Deal", "Predicted Deal"],
            yticklabels=["Actual No Deal", "Actual Deal"])
plt.title("Confusion Matrix")
plt.ylabel("Actual")
plt.xlabel("Predicted")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_error, mean_squared_error

time_df=df[['Original Air Date', 'US Viewership']].dropna()
time_df=time_df.drop_duplicates(subset=['Original Air Date'])

time_df['Original Air Date']=pd.to_datetime(time_df['Original Air Date'], errors='coerce')
time_df=time_df.dropna(subset=['Original Air Date'])

time_df=time_df.set_index('Original Air Date')

df_time_series=time_df['US Viewership'].dropna().sort_index()

if len(df_time_series) < 10:
  raise ValueError ("Insufficient data for time series forecasting.")

split_idx=int(len(df_time_series)*0.8)
train, test = df_time_series[:split_idx], df_time_series[split_idx:]

model=ExponentialSmoothing(train, trend="add", seasonal=None, damped_trend=True).fit()

predictions = model.forecast(len(test))

mae = mean_absolute_error(test, predictions)
mse = mean_squared_error(test, predictions)
rmse = np.sqrt(mse)

accuracy_metrics=pd.DataFrame({
    "Metric": ["Mean Absolute Error (MAE)", "Mean Squared Error (MSE)", "Root Mean Squared Error (RMSE)"],
    "Value": [mae, mse, rmse]
})

plt.figure(figsize=(10,5))
plt.plot(train.index, train, label="Train Data", color="blue")
plt.plot(test.index, test, label="Actual Test data", color="black")
plt.plot(test.index, predictions.values, label="Predicted", linestyle="--")
plt.legend()
plt.xlabel("Date")
plt.ylabel("US Viewership")
plt.title("Exponential Smoothing Model - US Viewership Forecasting")
plt.show()

print("\n===Accuracy Metrics===")
print(accuracy_metrics)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
import statsmodels.api as sm
import itertools
import pmdarima as pm # Import pmdarima

time_df=df[["Original Air Date", "US Viewership"]].dropna()

time_df["Original Air Date"]=pd.to_datetime(time_df["Original Air Date"], errors='coerce')
time_df=time_df.drop_duplicates(subset=['Original Air Date'])
time_df=time_df.dropna(subset=["Original Air Date"])

time_df=time_df.set_index('Original Air Date').asfreq('D')

# Fill the NaNs introduced by asfreq to make the series continuous
# Using ffill() and bfill() or interpolate() is common for ARIMA models
df_time_series=time_df['US Viewership'].ffill().bfill().sort_index()

# Ensure no NaNs remain after filling (e.g., if series was entirely NaN initially)
if df_time_series.isnull().any():
    raise ValueError("Time series still contains NaNs after filling. Cannot proceed with ARIMA.")

if len(df_time_series)<10:
  raise ValueError("Insufficient data for time series forecasting.")

split_idx=int(len(df_time_series)*0.8)
train, test = df_time_series[:split_idx], df_time_series[split_idx:]

# Use auto_arima to find the best model
# max_p, max_d, max_q are set to be small to avoid excessive computation time,
# but can be increased if needed
best_model = pm.auto_arima(train,
                           start_p=0, start_q=0,
                           max_p=3, max_d=2, max_q=3, # Increased ranges slightly
                           seasonal=False,
                           d=None, # let auto_arima determine d
                           trace=True,
                           error_action='ignore',
                           suppress_warnings=True,
                           stepwise=True)

# Get the best order found by auto_arima
best_order = best_model.order

# Make predictions using the best model
predictions = best_model.predict(n_periods=len(test))
predictions = pd.Series(predictions, index=test.index)

mae = mean_absolute_error(test, predictions)
mse = mean_squared_error(test, predictions)
rmse = np.sqrt(mse)

print(f"Best ARIMA Order: {best_order}")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")

print("\n### Best ARIMA Model Summary ###")
print(best_model.summary())

plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Train Data', color='blue')
plt.plot(test.index, test, label='Actual Test Data', color='black')
plt.plot(test.index, predictions, label='ARIMA Predictions', color='red', linestyle='--')
plt.title('ARIMA Model - US Viewership Forecast')
plt.xlabel('Date')
plt.ylabel('US Viewership')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Define selected predictors and dependent variable
selected_features = [
'Original Ask Amount','Industry','Pitchers Gender','Multiple Entrepreneurs','Season Number','Original Offered Equity','Valuation Requested']
# Drop missing values for relevant columns
lazy_df = df[selected_features + ['Got Deal']].dropna().copy()

# Convert 'Got Deal' to binary (1 = Got a deal, 0 = No deal)
lazy_df['Got Deal'] = lazy_df['Got Deal'].astype(int)

# Ensure numeric columns are properly formatted
numeric_cols = ['Original Ask Amount', 'Multiple Entrepreneurs']
for col in numeric_cols:
    lazy_df[col] = pd.to_numeric(lazy_df[col], errors='coerce')

# Convert categorical variables to numeric using Label Encoding
label_encoders = {}
for col in lazy_df.select_dtypes(include=['object']).columns:
    label_encoders[col] = LabelEncoder()
    lazy_df[col] = label_encoders[col].fit_transform(lazy_df[col].astype(str))

# Drop rows with missing values (to avoid errors in training)
lazy_df = lazy_df.dropna()

# Classification Setup (Target: Got Deal)
X_class = lazy_df.drop(columns=['Got Deal'])
y_class = lazy_df['Got Deal']

# Standardize numerical features
scaler = StandardScaler()
X_class_scaled = scaler.fit_transform(X_class)

# Train-Test Split (80-20)
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class_scaled, y_class, test_size=0.2, random_state=42)

# Run LazyPredict for Classification
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models_classification, _ = clf.fit(X_train_class, X_test_class, y_train_class, y_test_class)

# Display results for Classification
print("\n=== LazyPredict Classification Results ===\n")
print(models_classification)
```
